```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/Copy of Social_Network_Ads.csv')
print(df.head())
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Age'],df['EstimatedSalary'])
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.title('Age vs Estimated Salary')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Age'],df['Gender'])
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('Age vs Gender')

plt.show()
```
